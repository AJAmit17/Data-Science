Q1. How can you create a Bokeh plot using Python code?

Answer: Bokeh is a Python library for creating interactive visualizations and plots in web browsers


```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_file("scatter.html")

p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, size=10, color="navy")
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


Answer: In Bokeh, we can customize the appearance of a plot in many ways, including changing the axis labels, adding a title, adjusting the font sizes, and adding a legend.

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_file("customization.html")

p = figure(title="Customization Example", x_axis_label="X-axis", y_axis_label="Y-axis", plot_width=500, plot_height=400, background_fill_color="#f5f5f5")

p.xaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_font_size = "14pt"

p.xaxis.major_label_text_font_size = "12pt"
p.yaxis.major_label_text_font_size = "12pt"

p.circle(x, y, size=15, color="navy", legend_label="Data Points")
p.line(x, y, line_width=2, color="red", legend_label="Line")

p.legend.location = "top_right"
p.legend.title = "Legend Title"
p.legend.title_text_font_size = "12pt"
p.legend.label_text_font_size = "10pt"

show(p)
```

Answer: A Bokeh server is a Python process that allows you to create and deploy interactive Bokeh plots that can be updated in real-time. Using a Bokeh server, you can create web applications that provide a user interface for interacting with and exploring data.

Example:

```python
from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Slider
from bokeh.layouts import column

output_file('modal.html')
source = ColumnDataSource(data=dict(x=[1, 2, 3, 4, 5], y=[2, 5, 3, 1, 4], size=[10, 10, 10, 10, 10]))

plot = figure(plot_height=400, plot_width=600, title="Scatter Plot Example", tools="reset,box_zoom,pan")

scatter = plot.scatter(x="x", y="y", source=source, size="size", color="navy", alpha=0.5)

def update_size(attrname, old, new):
    size = slider.value
    scatter.data_source.data['size'] = [size, size, size, size, size]

slider = Slider(start=5, end=50, value=10, step=5, title="Marker Size")

slider.on_change('value', update_size)
layout = column(slider, plot)

curdoc().add_root(layout)

show(plot)
show(slider)
show(layout)
```
