Q1 . Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer : <br> Barplot - Used to visualize the relationship between a categorical variable and a numeric variable. It displays the mean of the variable along with the confidence interval.

Scatterplot - Used to visualize the relationship between two numeric variables. It shows the distribution of the data points and helps to identify any patterns or outliers.

Boxplot - Used to display the distribution of a dataset by showing the median, quartiles, and outliers. It is useful for detecting and visualizing the differences in a variable across different groups. 

Heatmap - Used to display the correlation between two or more variables. It is a way to represent the data in a color-coded matrix. 

Lineplot - Used to display the relationship between two numeric variables with a line connecting the data points. It is useful for visualizing trends over time or across different groups

Q2 . Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

Q3 . Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

titanic_df = sns.load_dataset('titanic')

sns.boxplot(x='pclass', y='age', data=titanic_df)
sns.boxplot(x='pclass', y='fare', data=titanic_df)
```

Q4 . Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

sns.histplot(x='price', data=diamonds, hue='cut', kde=True)
```

Q5 . Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

iris  = sns.load_dataset('iris')

sns.pairplot(iris , hue ='species')
```

Q6 . Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

flights_data = sns.load_dataset("flights")

sns.heatmap(data=flights_data.pivot(index="month", columns="year", values="passengers"), cmap="coolwarm")
```
