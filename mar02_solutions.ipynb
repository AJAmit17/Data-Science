Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer :
Matplotlib is a popular data visualization library in Python. It allows users to create a wide range of visualizations, such as line plots, scatter plots, bar plots, histograms, and pie charts. It is used to create high-quality graphical representations of data to better understand and make sense of data.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line plot - displays data points connected by straight line segments

2. Scatter plot - displays points on a two-dimensional plane, where each point represents a single variable with two values

3. Bar plot - displays vertical or horizontal bars to represent data in different categories

4. Histogram - displays the distribution of a single variable by dividing it into a set of intervals and counting the number of observations in each interval.

5. Pie chart - displays the proportion of a categorical variable as a circular graph, with individual slices representing each category's share in the total data.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np<br>
np.random.seed(3)<br>
x = 3 + np.random.normal(0, 2, 50)<br>
y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Answer : <br> A scatter plot is a graphical representation of the relationship between two numerical variables. The data is plotted as a collection of points, where each point represents a single observation of the two variables. The position of each point in the plot is determined by the values of the two variables.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:<br>
import numpy as np<br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])<br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer : <br> The subplot() function is used to create multiple plots within the same figure. This can be useful to visualize different aspects of the data or to compare multiple datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

axs[0,0].plot(x1, y1)
axs[0,1].plot(x2, y2)
axs[1,0].plot(x3, y3)
axs[1,1].plot(x4, y4)

axs[0,0].set_title('Line 1')
axs[0,1].set_title('Line 2')
axs[1,0].set_title('Line 3')
axs[1,1].set_title('Line 4')

fig.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.<br>
import numpy as np<br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"])<br>
profit = np.array([3000, 8000, 1000, 10000])

Answer : <br> A bar plot is a graph that represents categorical data with rectangular bars where the height or length of the bar is proportional to the value being represented. It is useful for visualizing and comparing data between different categories.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit ($)")
plt.ylabel("Company")

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.<br>
box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)

Answer : <br> A box plot is a graphical representation of a set of data through their quartiles. It consists of a rectangular box marked by the first quartile (Q1) and the third quartile (Q3), which contains the middle 50% of the data. The line inside the box represents the median. The "whiskers" extend from the box to the minimum and maximum values of the data, excluding outliers which are plotted as individual points.

Box plots are used to provide a summarized view of the distribution of a dataset, especially when comparing multiple datasets. They highlight the degree of variability, skewness, and the presence of outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
fig, ax = plt.subplots()
ax.set_title('Box Plot')
ax.boxplot(data)
ax.set_xticklabels(['box1', 'box2'])

plt.show()
```
